```python
# Create correlation matrix
corr_mat = df.corr(method='pearson')

# Retain upper triangular values of correlation matrix and
# make Lower triangular values Null
upper_corr_mat = corr_mat.where(
    np.triu(np.ones(corr_mat.shape), k=1).astype(bool))

# Convert to 1-D series and drop Null values
unique_corr_pairs = upper_corr_mat.unstack().dropna()

# Sort correlation pairs
sorted_mat = unique_corr_pairs.sort_values()
print(sorted_mat.sort_values(ascending=False))
```

```python
from scipy.stats import tukey_hsd
res = tukey_hsd(df_1['leftElbow'], df_2['leftElbow'],
         df_3['leftElbow'], df_4['leftElbow'])
print(res)
```

```python
from scipy.stats import tukey_hsd
res = tukey_hsd(df_1['leftShould'], df_2['leftShould'],
         df_3['leftShould'], df_4['leftShould'])
print(res)
```

```python
from scipy.stats import tukey_hsd
res = tukey_hsd(df_1['leftWrist'], df_2['leftWrist'],
         df_3['leftWrist'], df_4['leftWrist'])
print(res)
```

Мы  можем вычислить доверительный интервал, связанный с выбранным нами уровнем достоверности.

```python
conf = res.confidence_interval(confidence_level=.99)
for ((i, j), l) in np.ndenumerate(conf.low):
    # filter out self comparisons
    if i != j:
        h = conf.high[i,j]
        print(f"({i} - {j}) {l:>6.3f} {h:>6.3f}")
```

```python
# Kruskal-Wallis H-test
from scipy.stats import kruskal


# compare samples
stat, p = kruskal(df_1['leftElbow'], df_2['leftElbow'],
         df_3['leftElbow'], df_4['leftElbow'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

```python
# Kruskal-Wallis H-test
from scipy.stats import kruskal


# compare samples
stat, p = kruskal(df_1['leftShould'], df_2['leftShould'],
         df_3['leftShould'], df_4['leftShould'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

```python
# Kruskal-Wallis H-test
from scipy.stats import kruskal


# compare samples
stat, p = kruskal(df_1['leftWrist'], df_2['leftWrist'],
         df_3['leftWrist'], df_4['leftWrist'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

Приближение Александера-Говерна проверяет равенство k независимых средних в условиях неоднородности дисперсии. Тест применяется к образцам из двух или более групп, возможно, с разными размерами.

Использование этого теста основано на нескольких предположениях.

Образцы являются независимыми.

Каждая выборка относится к нормально распределенной популяции.

В отличие от f_oneway, этот тест не предполагает гомоскедастичность, вместо этого ослабляя предположение о равных дисперсиях.

Входные выборки должны быть конечными, одномерными и иметь размер больше единицы.
