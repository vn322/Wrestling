```python
# Create correlation matrix
corr_mat = df_1.corr(method='pearson')
np.fill_diagonal(corr_mat.values, 0)
# Convert correlation matrix to 1-D Series and sort
sorted_mat = corr_mat.unstack().sort_values(ascending=False)

print(sorted_mat)
s =sorted_mat.abs()
s[s>0.5]
```

```python
# Create correlation matrix
corr_mat = df_1.corr(method='pearson')

# Retain upper triangular values of correlation matrix and
# make Lower triangular values Null
upper_corr_mat = corr_mat.where(
    np.triu(np.ones(corr_mat.shape), k=1).astype(bool))

# Convert to 1-D series and drop Null values
unique_corr_pairs = upper_corr_mat.unstack().dropna()

# Sort correlation pairs
sorted_mat = unique_corr_pairs.sort_values()
print(sorted_mat.sort_values(ascending=False))
```

# Статистика

```python
# Mann-Whitney U test

from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_1['AngleY(°)'], df_2['AngleY(°)'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

```python
# Mann-Whitney U test

from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_1['AngleZ(°)'], df_2['AngleZ(°)'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

```python
# Kruskal-Wallis H-test
from scipy.stats import kruskal


# compare samples
stat, p = kruskal(df_1['AngleX(°)'], df_2['AngleX(°)'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```
